# Percobaan **1**

FAKTA: parent(Orangtua, Anak)

```python
data_parent = [
    ("alya", "bima"),
    ("alya", "satria"),
    ("bima", "david"),
    ("bima", "emma"),
    ("satria", "yunita"),
    ("satria", "grace")
]
```

ATURAN: Sibling(Saudara Kandung)
X dan Y adalah saudara jika memiliki orang tua (Z) yang sama dan X != Y

```python
def get_siblings(target):
  parents = [p for p, c in data_parent if c == target]
  siblings = set()
  for parent in parents:
    children = [c for parent, c in data_parent if parent == parent and c != target]
    siblings.update(children)
  return list(siblings)
```

ATURAN: Grandparent (Kakek/Nenek)
X adalah kakek/nenek Y jika X orang tua Z, dan Z orang tua Y

```python
def get_grandparents(target_cucu):
  results=[]
  # Cari orang tua dari target_cucu
  parents = [p for p, c in data_parent if c == target_cucu]
  for p in parents:
    # Cari orang tua dari orang tua (kakek/nenek)
    grandparents = [gp for gp, child in data_parent if child == p]
    results.extend(grandparents)
  return results
```

Uji Coba Query

```python
print(f"Saudara Bima: {get_siblings('bima')}")
print(f"Kakek/Nenek Emma: {get_grandparents('emma')}")
```

# Percobaan 2

Database: penyakit dan daftar gejalanya

```python
rules_penyakit = {
    "Malaria Tertiana": {"nyeri_otot", "muntah", "kejang"},
    "Malaria Quartana": {"menggigil", "tidak_enak_badan", "nyeri_otot"},
    "Malaria Tropika": {"keringat_dingin", "sakit_kepala", "mimisan", "mual"},
    "Malaria Pernisiosa": {"menggigil", "tidak_enak_badan", "demam", "mimisan", "mual"}
}
```

```python
def diagnosa_malaria(gejala_input):
  hasil_diagnosa = []

  for penyakit, gejala_syarat in rules_penyakit.items():
    # Cek apakah SEMUA gejala syarat terpenuhi oleh input user
    if gejala_syarat.issubset(gejala_input):
      hasil_diagnosa.append(penyakit)
  return hasil_diagnosa if hasil_diagnosa else ["Tidak terdeteksi penyakit"]
```

```python
#Simulasi seperti perintah assertz di prolog
gejala_pasien={"nyeri_otot", "menggigil", "tidak_enak_badan"}
print(f"Hasil Diagnosa: {diagnosa_malaria(gejala_pasien)}")
```

```python
# Working memory
gejala_pasien = []

def tanya_gejala(kode_gejala, teks_pertanyaan):
  # Menampilkan pertanyaan (Simulasi predikat pertanyaan/1 di Prolog)
  jawaban = input(f"{teks_pertanyaan} (y/t): ").lower()

  # Jika jawaban 'y', masukkan ke dalam list gejala (Simulasi assertz)
  if jawaban == 'y':
    gejala_pasien.append(kode_gejala)
```
